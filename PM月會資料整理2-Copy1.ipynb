```python
import os
import pandas as pd
import numpy as np
##基本內容設定
# folder_path = "/Users/winniecheng/Desktop"
# in_path = os.path.join(folder_path, '202112 PM Monthly review.xlsx')
input_path = '02112 PM Monthly review.xlsx'
output_path = os.path.join(folder_path, '市場資料_202112.xlsx')
##欄位設定
melt_col_list = ['Operator', 'Channel', 'Brand', 'Model Name(Daily Report)', 'Model Name', 
                     'SRP', 'Price Tier\nNEW', '頻段']
final_col_list = ['Operator', 'Channel', 'Brand', 'FET_model_name', 'out_model_name', 
                     'SRP', 'Price Tier/NEW', '頻段', '商品銷售日', '銷售量']
in_sheet = '工作表1'
##Model name彙整，use all_name
def model_combine_fuc(all_name, in_brand):
    if all_name is np.nan:
        return 'type error'
#     elif 'Samsung' in all_name:
#         return str(all_name).re.sub('[0-2]', '')
    else:
        return str(all_name).replace(in_brand + ' ', '')
##時間清理
def time_clean_fuc(tmp_time):
    tmp_date_list = tmp_time.replace(" ", '').split("'")
    tmp_m = tmp_date_list[0]
    tmp_y = tmp_date_list[1]
    ##時間重組
    y_v = '20' + str(int(tmp_y))
    m_v = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'].index(tmp_m)
    return str(y_v) + '/' + str(m_v + 1) + '/01' 
##主程式
if __name__ == '__main__':
    ##檔案讀取
    df = pd.read_excel(in_path, sheet_name = in_sheet)
    ##刪除不需要price tier column
    df = df.drop(columns=['Price Tier'])
    ##選定欄位範圍
    col_list = df.columns.tolist()
#     col_list = col_list[:col_list.index("MoM")]
    df = df[col_list]
    ##將資料整理成匯出形式
    out_df = pd.melt(df, id_vars=melt_col_list)
    out_df.columns = final_col_list
    out_df['new_model_name'] = out_df.apply(lambda x: model_combine_fuc(x.out_model_name, x.Brand), axis=1)
    out_df['商品銷售日'] = out_df['商品銷售日'].apply(time_clean_fuc)
    ##正則式取代()內資訊
    out_df['new_model'] = out_df.new_model_name.str.replace('\([a-bA-b0-9\+\)/ ]+', '', regex=True)
    ##檔案儲存
    writer = pd.ExcelWriter(output_path, engine = 'xlsxwriter')
    out_df.to_excel(writer, index=False, sheet_name='總表new')
    writer.save()
```
